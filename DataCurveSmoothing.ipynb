# **Curve Smoothing**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.interpolate import make_interp_spline
```

```python
np.random.seed(42)
```

```python
x = np.random.permutation(np.arange(1,20000))[:100]
x = np.sort(x)
```

```python
y = np.random.permutation(np.arange(-3000,3000))[:100]
```

```python
X_Y_spline = make_interp_spline(x,y)

x_interp = np.linspace(x.min(),x.max(),1000) # generate 500 sample within min - max
y_interp = X_Y_spline(x_interp)
```

```python
plt.figure(figsize=(15,7))
plt.plot(x,y)
plt.title('Before: Curve Smoothing')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(x_interp,y_interp)
plt.title('After: Curve Smoothing')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```
